```python
import pandas as pd
import numpy as np
df = pd.read_csv('BowlingSummaryINDIA.csv')
# df.tail()
```

```python
df = df.drop(['Ab'],axis=1)
# print(df)
```

```python
df
```

```python
df['Matches Played'] = df.groupby('Name')['Name'].transform('count')
```

```python
df
```

```python
grouped = df.groupby('Name')
aggregated = grouped[['Overs','Maiden','Runs','Wickets','0s',"4s","6s",'Wide','NB']].sum()
aggregated
```

```python
df = df.sort_values(by='Name')
df
```

```python
df = df.drop_duplicates(subset=['Name'])
df
```

```python
df1 = df[['Name', 'Matches Played']]
df1
# df1.to_csv('Matches Played.csv')
```

```python
df['Runs']=list(aggregated['Runs'])
df['Overs']=list(aggregated['Overs'])
df['Maiden']=list(aggregated['Maiden'])
df['Wickets']=list(aggregated['Wickets'])
df['0s']=list(aggregated['0s'])
df['4s']=list(aggregated['4s'])
df['6s']=list(aggregated['6s'])
df['Wide']=list(aggregated['Wide'])
df['NB']=list(aggregated['NB'])
df['ECON']=df['Runs']/df['Overs']
df
```
